## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'iris_dataset.csv'</i>.

```python
# Load the data
data = pd.read_csv("iris_dataset.csv")
data
```

## Plot the data

For this exercise, try to cluster the iris flowers by the shape of their sepal. 

<i> Use the 'sepal_length' and 'sepal_width' variables.</i> 

```python
# Create a scatter plot based on two correspoding features (spel_length and sepal_width; OR petal_length and petal_width)
plt.scatter(data["sepal_length"], data["sepal_width"])
plt.xlabel("Length of sepal")
plt.ylabel("Width of sepal")
plt.show()
```

# Clustering (unscaled data)

Separate the original data into 2 clusters.

```python
# Create a variable which will contain the data for the clustering
x = data.copy()
# Create a k-means object with 2 clusters 
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
# Create a copy data, so we can see the clusters next to the originam data
clusters = data.copy()
# Predict the cluster for each observation
clusters["Cluster_pred"] = kmeans.fit_predict(x)
```

```python
# Create a scatter plot based on two corresponding features
plt.scatter(data["sepal_length"], data["sepal_width"], c=clusters["Cluster_pred"], cmap="rainbow")
plt.xlabel("Length of sepal")
plt.ylabel("Width of sepal")
plt.show()
```

# Standardize the variables

Import and use the <i> method </i> function from sklearn to standardize the data. 

```python
# import some preprocessing module
from sklearn import preprocessing

# Scale the data for better results
x_scaled = preprocessing.scale(data)
x_scaled
```

# Clustering (scaled data)

```python
# Create a k-means object with 2 clusters
kmeans_scaled = KMeans(2)
# Fit the data
kmeans_scaled.fit(x_scaled)
```

```python
# Create a copy of data, so we can see the clusters next to orginal data
clusters_scaled = data.copy()
# Predict the cluster for each observation
clusters_scaled["clusters_pred"] = kmeans_scaled.fit_predict(x_scaled)
```

```python
clusters_scaled
```

```python
# Create a scatter plot based on two correspoding features
plt.scatter(clusters_scaled['sepal_length'], clusters_scaled['sepal_width'], c= clusters_scaled ['clusters_pred'], cmap = 'rainbow')
plt.xlabel("Length of sepal")
plt.ylabel("Width of sepal")
plt.show()
```

## Take Advantage of the Elbow Method

### WCSS

```python
wcss = []
for i in range(1, 10):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss_iner = kmeans.inertia_
    wcss.append(wcss_iner)
wcss
```

### The Elbow Method

```python
number_cluster = range(1, 10)
plt.plot(number_cluster, wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```

```python
# 4 Cluster
kmeans = KMeans(4)
kmeans.fit(x_scaled)
clusters_scaled_4 = kmeans.fit_predict(x_scaled)
plt.scatter(data["sepal_length"], data["sepal_width"], c=clusters_scaled_4, cmap="rainbow")
```

```python
# 5 Cluster
kmeans = KMeans(5)
kmeans.fit(x_scaled)
clusters_scaled_5 = kmeans.fit_predict(x_scaled)
plt.scatter(data["sepal_length"], data["sepal_width"], c=clusters_scaled_5, cmap="rainbow")
```
