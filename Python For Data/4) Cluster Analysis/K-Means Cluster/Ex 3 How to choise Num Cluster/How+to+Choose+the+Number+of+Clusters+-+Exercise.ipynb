## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Countries_exercise.csv'</i>.

```python
# Load the data
raw_data = pd.read_csv("Countries_exercise.csv")
raw_data
```

Remove the duplicate index column from the dataset. 

```python
data = raw_data.copy()
```

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

## Plot the data

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

```python
plt.scatter(data["Longitude"], data["Latitude"])
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

## Select the features

Make sure to select the appropriate features since we are no longer using the categorical variable for our clustering but rather <i>'Longitude'</i> and <i>'Laditude'</i>.

```python
x = data.iloc[:, 1:3]
# Mean: choise all row, coloumn 1 and 2
x
```

## Clustering

Use 4 clusters initially.

```python
kmeans = KMeans(4)
kmeans.fit(x)
```

### Clustering results

```python
indentified_clusters = kmeans.fit_predict(x)
indentified_clusters
```

```python
data_with_cluster = data.copy()
data_with_cluster["Cluster"] = indentified_clusters
data_with_cluster
```

Plot the data once again but separate the data by the clusters we defined.  

## Plot the data

```python
plt.scatter(data["Longitude"], data["Latitude"], c=indentified_clusters, cmap='rainbow')
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

## Selecting the number of clusters

### WCSS

Use the ingerated <i>sklearn</i> method <i> 'inertia_' </i>.

```python
kmeans.inertia_
```

Write a loop that calculates and saves the WCSS for any number of clusters from 1 up to 10 (or more if you wish).

```python
wcss = []
# "cl_num" is a that keeps track the highest number of clusters we want to use the WCSS method for.
# Note that "range" doesn't include the upper boundery
cl_num = 11
for i in range(1, cl_num):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
```

```python
wcss
```

### The Elbow Method

```python
number_clusters = range(1, cl_num)
plt.plot(number_clusters, wcss)
plt.title("The Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("Within-cluster Sum of Squares")
```
